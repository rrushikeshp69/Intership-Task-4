# **Load and Clean the Data**

```python
# Import the pandas library
import pandas as pd

# Load the CSV file into a pandas DataFrame.
# The 'encoding' helps prevent errors with special characters.
df = pd.read_csv('ecommerce.csv', encoding='ISO-8859-1')
print("File loaded successfully!")

# --- Data Cleaning ---

# 1. Combine 'Order_Date' and 'Time' into a single datetime column
df['DateTime'] = pd.to_datetime(df['Order_Date'] + ' ' + df['Time'])

# 2. Drop rows that don't have a Customer_Id
df.dropna(subset=['Customer_Id'], inplace=True)

# 3. Ensure key columns are the correct numeric type
df['Sales'] = pd.to_numeric(df['Sales'])
df['Quantity'] = pd.to_numeric(df['Quantity'])

# --- Display the first 5 rows to see our cleaned data ---
print("\nHere's a preview of the cleaned data:")
df.head()
```

# **Task 1: Find All Purchases for a Specific Customer**

```python
# Find all transactions for a sample customer (e.g., Customer_Id 1005)
# You can change the ID to any customer from your data.
customer_purchases = df[df['Customer_Id'] == 1005].sort_values(by='DateTime', ascending=False)

# Display the results
customer_purchases.head()
```

# **Task 2: Calculate Total Sales by Product Category**

```python
# Group data by product category and sum the sales, then sort it
sales_by_category = df.groupby('Product_Category')['Sales'].sum().sort_values(ascending=False)

# Display the top 10 categories by sales
sales_by_category.head(10)
```

# **Task 3: Find Top Selling Products by Quantity**

```python
# Group by product name and sum the quantity to find the most sold items
top_products_by_quantity = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)

# Display the top 10 best-selling products
top_products_by_quantity.head(10)
```
